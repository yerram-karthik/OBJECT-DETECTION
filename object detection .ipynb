```python
import cv2
import matplotlib.pyplot as plt
```

```python
frozen_model = "C:/Users/yerra/Downloads/OBJECT DETECTION/CODING/back_end/frozen_inference_graph.pb"
config_file = "C:/Users/yerra/Downloads/OBJECT DETECTION/CODING/back_end/ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt"
```

```python
model = cv2.dnn_DetectionModel(frozen_model, config_file)
```

```python
classLables=[]
file_name="C:/Users/yerra/Downloads/OBJECT DETECTION/CODING/back_end/labels.txt"
with open(file_name,'rt') as fpt:
    classLabels=fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

```python
print(len(classLabels))
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5,127,5,127.5))
model.setInputSwapRB(True)
```

```python
img=cv2.imread("C:/Users/yerra/Downloads/OBJECT DETECTION/CODING/back_end/boy.jpeg")
plt.imshow(img)
plt.show()
```

```python
ClassIndex,confidece,bbox=model.detect(img,confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
font_scale=1
font=cv2.FONT_HERSHEY_PLAIN
for ClassInd,conf,boxes in zip(ClassIndex.flatten(),confidece.flatten(),bbox):
    cv2.rectangle(img,boxes,(255,0,0),1)
    cv2.putText(img,classLabels[ClassInd-1],(boxes[0]+4,boxes[1]+20),font,fontScale=font_scale,color=(0,255,0,),thickness=0)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
